```python
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
```

## Tạo Custom Dataset
Một custom dataset cần 3 hàm : __init__, __len__, __get_item__

```python
f = open("news_categories.txt","r",encoding="utf-8")

label = []
text = []
for i in f:
    label.append(i.split()[0])
    text.append(" ".join(i.split()[1:]))

X_train, X_test, y_train, y_test = train_test_split(text, label, test_size=0.2)

print(len(X_train))
print(len(X_test))
```

```python
class Dataset_custom(Dataset):
    #Hàm khởi tạo
    def __init__(self, label, text):
        self.label = label
        self.text = text
    
    #Return số lượng bản ghi
    def __len__(self):
        return len(self.text)

    #load và return mẫu dữ liệu tại index 
    def __getitem__(self, index):
        return self.text[index], self.label[index] 
    
```

```python
training_set = Dataset_custom(label=y_train, text=X_train)
test_set = Dataset_custom(label=y_test, text=X_test)
```

```python
train_iter = DataLoader(training_set, batch_size=64, shuffle=True)
test_iter = DataLoader(test_set, batch_size=64, shuffle=True)

print(len(train_iter))
print(len(test_iter))
```

```python
#load mẫu dữ liệu trong dataloader
train_features, train_labels = next(iter(train_iter))

print(train_features[0])
print(train_labels[0])
```

```python
for batch, (X, y) in enumerate(train_iter):
    print(batch)
    print(len(X))
    print(len(y))
```
